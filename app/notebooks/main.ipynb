```python
import requests
from bs4 import BeautifulSoup

#Passo a URL e coleto a lista de vagas de cientista de dados no Brasil
url = 'https://www.linkedin.com/jobs/search?keywords=Cientista%20De%20Dados&location=Rio%20de%20Janeiro%2C%20Rio%20de%20Janeiro%2C%20Brasil&geoId=106701406&distance=0&f_TPR=&position=1&pageNum=0'
response = requests.get(url)
if response.status_code == 200:
   soup = BeautifulSoup(response.text, 'html.parser')
else:
   print("Erro ao acessar a lista de vagas")
```

```python
#Pego somente os cards de vagas
job_listings = soup.find_all('div', {'class':'job-search-card'})

links_vagas = []
#Dentro de cada card, pego o link da vaga
for job in job_listings:
      tag = job.find('a', class_='base-card__full-link')
      links_vagas += tag['href']
      print(links_vagas)
```
